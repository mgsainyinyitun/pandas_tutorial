## Scatter Plot

```python
# 1. Create or prepare for data
x = np.array([np.random.randint(0,100) for i in range(100)])
y = np.array([np.random.randint(0,100) for i in range(100)])

x1 = np.array([np.random.randint(0,100) for i in range(100)])
y1 = np.array([np.random.randint(0,100) for i in range(100)])
```

```python
# 2. Create Figure
fig,axis = plt.subplots(figsize=(10,5))
# 3. Plot
axis.scatter(x,y,c="#C30CBA",marker="x")
axis.scatter(x1,y1,c="#34C30C",marker="o",s=100)
# 4. Customize
axis.set_title("Sample Scatter")
axis.set_xlabel("X-label")
axis.set_ylabel("Y-label")
```

## Bar Plot

```python
# 1. Create Data
x=[1,2,3,4,5,6,7]
height = [130,200,110,10,39,99,70]
height2 = [30,800,10,5,29,19,20]

# 2. Create Plot
fig,axes = plt.subplots(figsize=(10,5))

# 3. Plot
axes.bar(x,height,color="#503199",width=0.5)
axes.bar(x,height2,color="#09E7F0",width=0.5)

# customize
axes.set_title("Bar Sample")
```

## Histogram Plot

```python
# 1. Create Data
mu,sig = 120,15; # standard devi, variance
#x = mu + sig * np.random.randn(10000) # ( mu*sig*x )
x = np.random.normal(120,60,(10000))

# 2. Create fig
fig,axes = plt.subplots()

# 3. plot
axes.hist(x,50,density=True,facecolor='g', alpha=0.75)

# 4. Customize
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.xlim(0, 200)
#plt.ylim(0, 0.03)
plt.grid(True)
axes.set_title('Sample Histogram Plot')
```

## Subplot Option 1 

```python
# 1. Create Data
x = np.linspace(0,20,30)
y = np.sin(x)

x1 = np.random.normal(120,60,(10000))

# 2. create figure
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5));

# 3. Plot
ax1.plot(x,y,'r')
ax2.hist(x1,30)
ax3.plot(x,y,'g')
ax4.hist(x1,50)

# 4. customize

ax1.set_title("Sample Sine wave")
ax2.set_title("Sample Histogram")
```

```python
## Subpot Option 2

# 1. Create Data
x = np.linspace(0,20,30)
y = np.sin(x)

x1 = np.random.normal(120,60,(10000))

# 2. create figure
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5));

# 3. Plot
ax[0,0].plot(x,y)
ax[0,1].hist(x1,50)
ax[1,0].plot(x,y)
ax[1,1].hist(x1,50)

# 4. customize
#plt.tight_layout.auto_adjust_subplotpars(fig)
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=3.0)

ax[0,0].set_title("Sample Sine wave")
ax[0,1].set_title("Sample Histogram")
ax[1,0].set_title("Sample Sine wave")
ax[1,1].set_title("Sample Histogram")
```
